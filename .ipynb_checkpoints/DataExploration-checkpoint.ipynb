```python
import pandas as pd
import config
import mysql.connector 
from mysql.connector import errorcode
from sqlalchemy import create_engine
```

```python
#Create connection point to database
engine = create_engine("mysql+mysqlconnector://{user}:{pw}@{host}/{db}".format(
        user = config.user,
         pw= config.passwd,
         host = config.host,
         db = config.db_name))

#Retreive data
data = pd.read_sql('select * from okquakes;',con=engine)
```

```python
#Data cleanup
data['origintime'] = pd.to_datetime(data['origintime'])
data = data.mask(data.eq('None')).dropna()
data = data.drop(columns=['index'])
data = data.reset_index()
```

```python
#Plot seismic events per year
data['origintime'].groupby([pd.DatetimeIndex(data.origintime).year]).agg('count').plot()
data['magnitude'] = data['magnitude'].astype(float)
bigs = data[data['magnitude']>2]
bigs['origintime'].groupby([pd.DatetimeIndex(bigs.origintime).year]).agg('count').plot()
```
